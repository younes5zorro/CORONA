```python
import warnings
import pandas as pd
import re
import itertools
import codecs
import json
import os

warnings.filterwarnings('ignore')
```

```python
dialectid_data=pd.read_csv('data/dialectid_data.tsv',delimiter='\t',encoding='utf-8',error_bad_lines=False)
dialectid_data
```

```python
df = (pd.concat([pd.DataFrame([[row['Sentence1'],row['DClass1']],
                         [row['Sentence2'],row['DClass2']],
                         [row['Sentence3'],row['DClass3']],
                         [row['Sentence4'],row['DClass4']],
                         [row['Sentence5'],row['DClass5']],
                         [row['Sentence6'],row['DClass6']],
                         [row['Sentence7'],row['DClass7']],
                         [row['Sentence8'],row['DClass8']],
                         [row['Sentence9'],row['DClass9']],
                         [row['Sentence10'],row['DClass10']],
                         [row['Sentence11'],row['DClass11']],
                         [row['Sentence12'],row['DClass12']]
                        ],columns=['sentence','class'])  
           
           for _, row in dialectid_data.iterrows()],ignore_index=True).reset_index()).dropna()
```

```python
df = df.drop_duplicates(subset=['sentence'])
```

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)

def clear(sentence):
    sentence = emoji_pattern.sub(r'',sentence)
    sentence = re.sub(r'^RT', '', sentence)
    sentence = re.sub(r'\&\w*;', '', sentence)
    sentence = re.sub(r'\@\w*', '', sentence)
    sentence = re.sub(r'\$\w*', '', sentence)
    sentence = re.sub(r'https?:\/\/.*\/\w*', '', sentence)
    sentence = re.sub(r'#\w*', '', sentence)
    sentence = re.sub(r'[.]+', '', sentence)    
    sentence = re.sub(r'[\"\n\t,\.!?()-_$*&؟\"\\\'،%~]', '', sentence)    
    sentence = re.sub(r'[»]', ' ', sentence)

    sentence = sentence.lower()
    sentence = sentence.strip(' ')
    sentence = re.sub('  ', ' ', sentence)
    sentence = ''.join(ch for ch, _ in itertools.groupby(sentence))
    return sentence
```

```python
df['sentence'] = df['sentence'].apply(clear)
```

```python
df.to_csv("data_1.csv", sep=';', encoding='utf-8')
```

```python
import random

sentances = []
classes = []
dm = pd.DataFrame([])

list_dir = {"maghrebi":"mag_words.txt","tunisien":"tun_words.txt","general":"msa_words.txt","levantine":"lev_words.txt","gulf":"glf_words.txt","egyptian":"egy_words.txt"}

for key in list_dir:
    word_list = []
    
    with open(list_dir[key],encoding='utf-8') as f:  # or whatever the wordlist is saved as
        for line in f.readlines():
            word = line.strip()
            word_list.append(word)

        for i in range(5000):
            sentances.append(" ".join([word_list[random.randrange(0, len(word_list))] for i in range(random.randrange(1, 25))]))
            classes.append(key)

dm['sentence'] = sentances
dm['class'] = classes

dm.head()
```

```python
dm.to_csv("data_3.csv", sep=';', encoding='utf-8')
```

```python
import os

list_dir = {"levantine":"files_sy"}
dt = pd.DataFrame([])
sentances = []
classes = []
for key in list_dir:
    dirname = os.getcwd()+"\\"+list_dir[key]+"\\"
    files = os.listdir(os.getcwd()+"\\"+list_dir[key])
    for filename in files :
        with open(dirname+filename,'r',encoding='utf8') as f:
            for line in f.readlines():
                sentances.append(clear(line))
                classes.append(key)

dt['sentence'] = sentances
dt['class'] = classes

dt.to_csv("data_full_8.csv", sep=';', encoding='utf-8')
dt.head()
```
